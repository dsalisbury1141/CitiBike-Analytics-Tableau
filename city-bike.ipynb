```python
# Importing Dependencies
import pandas as pd
import csv
import time
import os
from datetime import date, timedelta
```

```python
#Defining source file path
Sept2019 = os.path.join("Resources","Sept2019-citibike-tripdata.csv")
Sept2020 = os.path.join("Resources","Sept2020-citibike-tripdata.csv")

#Reading the files into Pandas
sept19 = pd.read_csv(Sept2019)
sept20 = pd.read_csv(Sept2020)
```

```python
#updating gender field from numerical to actual
sept2019 = sept19.replace({"gender": {0: "Unknown", 1: "Male", 2: "Female"}})
sept2020 = sept20.replace({"gender": {0: "Unknown", 1: "Male", 2: "Female"}})
```

```python
#created new "age" field then calculated the age by subtracting the listed birth year and current year
sept2019['age'] = 2020 - sept2019['birth year']
sept2020['age'] = 2020 - sept2020['birth year']
```

```python
#sort file 2019 reviewing data to scrub
sept2019.sort_values('tripduration')
```

```python
#Removing all rows with < 120 seconds trip durationa and show the same station ID (no one can ride and come back in 2 mins)
sept2019_citibike = sept2019[(sept2019['tripduration'] > 120) | (sept2019['start station id'] != sept2019['end station id'])]

#Sort values and keep all tripduration < 12o with different station ID's remain
sept2019_citibike.sort_values('tripduration')
```

```python
#Removing all rows with < 120 seconds trip durationa and show the same station ID (no one can ride and come back in 2 mins)
sept2020_citibike = sept2020[(sept2020['tripduration'] > 180) | (sept2020['start station id'] != sept2020['end station id'])]

#Sort values and keep all tripduration < 12o with different station ID's remain
sept2020_citibike.sort_values('tripduration')
```

```python
#Cleaned dataset saved as .csv
sept2019_citibike.to_csv('Resources/sept2019.csv', index=False)
```

```python
#Cleaned dataset saved as .csv
sept2020_citibike.to_csv('Resources/sept2020.csv', index=False)
```
